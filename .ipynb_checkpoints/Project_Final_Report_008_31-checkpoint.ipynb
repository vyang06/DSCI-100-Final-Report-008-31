In this project, we investigate a real dataset collected by a research group in the Computer Science department at UBC. In this report, our group has chosen to address the following specific question:

**Can a player's age, gender, experience level, total playtime, number of sessions, and average session duration be used to predict whether they subscribe to the newsletter or not?**

Being able to predict this provides insight into which groups of players show more engagement with the project, which can help guide future outreach and recruitment strategies.

## Method & Results

### Method Description

### Wrangling Data

**1. Join datasets**
- Combine players.csv and sessions.csv using hashedEmail, we do this to find the number of sessions played by each individual player.

**2. Clean data**
- Cleaning the data removes or ignores any cells that contain a missing value (for example, missing Age or end_time).
  
**3. Feature creation**
- This is done to compute the total number of sessions and the average session duration for each player.
  
**4. Convert data types**
- Make sure subscribe is a logical (TRUE/FALSE) variable and categorical predictors use factor().
  
**5. Tidy data**
- Keep one row per player, with all relevant variables in columns.

```r
library(tidyverse)
library(tidymodels)
set.seed(2024)
options(repr.plot.width = 10, repr.plot.height = 7)

players <- read_csv("players.csv")
sessions <- read_csv("sessions.csv")
head(sessions)
head(players)
```

```r
combined_data <- bind_rows(
  players |> select(hashedEmail, subscribe, Age, played_hours),
  sessions |> mutate(session_duration_min = (original_end_time - original_start_time) / (1000 * 60)) |>
           select(hashedEmail, session_duration_min)
) |>
  group_by(hashedEmail) |>
  summarize(
    subscribe = first(subscribe),
    Age = first(Age),
    played_hours = first(played_hours),
    n_sessions = sum(!is.na(session_duration_min)),
    mean_duration_min = mean(session_duration_min, na.rm = TRUE)
  ) |>
  mutate(subscribe = as_factor(subscribe)) |>
  select(subscribe, Age, played_hours, n_sessions, mean_duration_min) |>
filter(n_sessions > 0, mean_duration_min > 0)


head(combined_data)
```

```r
mean_value <- combined_data |>
    summarize(mean_age = mean(Age, na.rm = TRUE), mean_played_hours = mean(played_hours, na.rm = TRUE)) |>
    pivot_longer(cols = c(mean_age, mean_played_hours), names_to = "Variable", values_to = "Mean")

mean_value
```

### Visualizations

```r
player_visualization_1 <- combined_data |> ggplot(aes(x = Age)) +
    geom_histogram(binwidth = 2, color = "white", na.rm = TRUE) +
    labs( title = "Distribution of Player Ages", x = "Age (years)", y = "Number of players") +
    theme(text = element_text(size = 16))

player_visualization_1
```

The graph above visualizes the amount of players for each age starting from 10 years to about 60 years of age. It shows that the majority of the MineCraft players are people between the ages of 15 and 25 years old. 

```r
player_visualization_2 <- ggplot(combined_data, aes(x = Age, fill = subscribe)) +
    geom_bar(position = "fill") +
    scale_y_continuous(labels = scales::percent_format()) +
    labs(title = "Proportion of Subscribed Players by Age", x = "Age", y = "Proportion") +
    theme(text = element_text(size = 15))

player_visualization_2
```

This graph shows the proportions of players that are subscribed or not for each age group. It shows that the age range that had the most players, 15 to 25 years old, are the people who are also most likely to not subsribed to the newsletter. Therefore, it can be concluded that age might not play a very big factor on deciding whether or not the player will subsribe. The stacked bar graph, allows for a very simple proportion visualization. 

```r
player_visualization_3 <- ggplot(combined_data, aes(x = gender, fill = subscribe)) +
    geom_bar(position = "fill") +
    labs(title = "Proportion of Subscribed Players by Gender", x = "Gender", y = "Proportion of players") +
    theme(text = element_text(size = 15))

player_visualization_3
```

This graph is in the same style as the one above it, except this one shows the proportions of players who subsribe based on their *gender*. Their is a pretty even distribution amongst the genders and their likelihood to subscribe to the news letter, so this variable might also not be the best in predicting the probability of players to subsribe to the newsletter. 

```r
player_visualization_4 <- 
    ggplot(combined_data, aes(x = played_hours, fill = subscribe)) +
    geom_histogram(position = "dodge") +
    labs(title = "Play Hours by Subscription", x = "Hours Played", y = "Count") +
    theme(text = element_text(size = 15))

player_visualization_4
```

This graph is also intended to compare a variable, specifically *hours played*, to the subscription status of the corresponding player. This visual shows that the smaller amount of hours played by an individual shows a lower likelihood in subscription to the newsletter. As the pattern is more distinct in this graph, we can say that using the variable, hours played is a good factor in predicting whether a player will subsribe or not. We chose to use a group/clustered bar chart as the range of hours played was massive and it was very hard to see the outcomes by using a stacked bar chart like the previous graphs. 

```r
library(tidyverse)
players_for_viz <- read_csv("players.csv")

#relationship between experience and played hours
options(repr.plot.width = 10, repr.plot.height = 6)
experience_vs_played_hours_plot <- players_for_viz |>
                                group_by(experience) |>
                                summarize(mean_hours = mean(played_hours, na.rm = TRUE)) |>
                                ggplot(aes(x = experience, y = mean_hours, fill = experience)) +
                                geom_col() +
                                labs(title = "Average Hours Played by Experience Level",
                                x = "Experience Level",
                                y = "Average Hours Played") +
                                theme(plot.title = element_text(hjust = 0.5, size = 20))
experience_vs_played_hours_plot
```

Players with a regular experience level have the highest played hours, this is relevant to liklihood of subscription if played hours is correlated to liklihood of subscription. 

```r
#relationship between gender and played hours
options(repr.plot.width = 10, repr.plot.height = 6)
gender_vs_played_hours_plot <- players_for_viz |>
                                group_by(gender) |>
                                summarize(mean_hours = mean(played_hours, na.rm = TRUE)) |>
                                ggplot(aes(x = gender, y = mean_hours, fill = gender)) +
                                geom_col() +
                                labs(title = "Average Hours Played by Gender",
                                x = "Gender",
                                y = "Average Hours Played") +
                                theme(plot.title = element_text(hjust = 0.5, size = 20))
gender_vs_played_hours_plot
```

Players identifying at non-binary have the highest played hours, this is relevant to liklihood of subscription if played hours is correlated to liklihood of subscription
